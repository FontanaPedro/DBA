```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler


import warnings
warnings.filterwarnings("ignore")
```

### Question 0
Load the telco data file we used in Lab 3, and perform a train-test split. Create a scaled version of the data, using Standard Scaler. Call the scaled version X_train_scaled, and X_test_scaled.


```python
#Reading and fast visualization of my dataset
df = pd.read_csv("3. telco.csv")
df.head()
```

```python

# I create my features values in y
X = df.drop(["Churn"], axis=1)

# I create my target value in X 
y=df["Churn"]


#Perfom a train-test split ; stratify by my target value so its equally in proportion in both sets and random_state=0 just to reproduce the same split in the future
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,random_state=0)

```

```python
# Creating the scaler variable
scaler=MinMaxScaler()

#the scaler needs to be trained only in the training data and not in the test data, so it needs to be fit in there.
# What this will do is to adapt to all the values from the training set and create the new parameters dividing 
# all the numbers by the Max number found in the trianig data set
scaler.fit(X_train)

# Once the scaler variable has all the relative numbers, i can transform all the features so my new X_train_scaled will 
# only have numbers potencially between 0-1. On the test data we might find numbers higher than one or just smaller since 
# it will be divided by the higher number in the trainig set
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test) 
```

<h2>Question 1</h2>
Build two versions of a default random forest model, using the normal data (X_train, X_test) and the scaled data (X_train_scaled, X_test_scaled). Print train and test scores for normal and scaled. </i>

```python

# Default random forest model --> n_estimators = 10 ; max_depth = None ; max_features=”sqrt” and
# will state random_state=0 to reproduce the difference between nomal data and scaled data

#I create the classifier and fit it in the NON-scaled data as the first point of comparisson
rfclf = RandomForestClassifier(random_state=0).fit(X_train,y_train)

#Save the scores of the NON-scaled data 
normal_train_score=rfclf.score(X_train,y_train)
normal_test_score=rfclf.score(X_test,y_test)

print("The score for the NON-scaled data is:  Train score: {:.3f}  ; Test score: {:.3f}\n".format(normal_train_score,normal_test_score))


##I create the classifier and fit it in the SCALED data to compare with the first one. While fitting with the SCALED data, the target is not scaled
rfsclf = RandomForestClassifier(random_state=0).fit(X_train_scaled,y_train)

#Save the scores of the SCALED data 
rfsclf_train_score=rfsclf.score(X_train_scaled,y_train)
rfsclf_test_score=rfsclf.score(X_test_scaled,y_test)

print("The score for the SCALED data is:  Train score: {:.3f}  ; Test score: {:.3f}\n".format(rfsclf_train_score,rfsclf_test_score))


###################### I SHOULD PLOT THE IN A GRAPH !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!


```

<h2>Question 2</h2>
Build two versions of a default mlp model, using the normal data (X_train, X_test) and the scaled data (X_train_scaled, X_test_scaled). Print train and test scores for normal and scaled. </i>

```python
# Defaul MLP model --> hidden_layers_size=100, activation function = "relu" ; alpha = 0.0001

#I create the mlp with dafault parameters for the NON-scaled data
mlpclf = MLPClassifier(random_state=0).fit(X_train, y_train)

#Save the scores of the NON-scaled data 
mlp_train_score=mlpclf.score(X_train,y_train)
mlp_test_score=mlpclf.score(X_test,y_test)


print ("The score for the NON-scaled data is:  train score: {:.3f} ;  test score: {:.3f} \n".format(mlp_train_score,mlp_test_score))


#I create the mlp with dafault parameters for the SCALED data
mlpsclf = MLPClassifier(random_state=0).fit(X_train_scaled, y_train)

#Save the scores of the NON-scaled data 
mlps_train_score=mlpsclf.score(X_train_scaled,y_train)
mlps_test_score=mlpsclf.score(X_test_scaled,y_test)


print ("The score for the NON-scaled data is:  train score: {:.3f} ;  test score: {:.3f} \n".format(mlps_train_score,mlps_test_score))
```

<h2>Question 3</h2>
Now try mlp models with different values for the hidden layer size. Try with  two hidden layers, with sizes ranging from 1 to 20, incremented by 5. For each iteration, fit the model to both the normal and scaled data. Save the train and test scores for both in two lists. Also print the scaled and normal results at each iteration.

```python
# MLP with different values for the hidden layer size --> 1 SINGLE later many units inside 

# This is the list with the units of each different hidden layers
units = list(range(1,20,5))

#I create empty lists where i will save the results of MLP with 1 single layer with different hidden layer sizes
mlp1_train_score=list()
mlp1_test_score=list()

# j will be just a counter variable to distinguish each iteraction in position of a list
j=0


# Just printing a title
print(" For the MLP model fit in the NON-scaled data with a SINGLE layer and different hidden sizes the results are:\n") 

# A loop where the MLP classifier will be fit with the different units for the NON-scaled data
for i in units: 
        
        mlp1clf=MLPClassifier(hidden_layer_sizes=i, random_state=0).fit(X_train,y_train)

        # I add to the list each of the scores of the iteration

        mlp1_train_score.append(mlp1clf.score(X_train,y_train))
        mlp1_train_score.append(mlp1clf.score(X_test,y_test))
        

        #While looping printing the results 
        print("{}) Hidden size: {}  --> Train Score: {:.3f}   ;   Test Score: {:.3f}".format(j+1,i, mlp1_train_score[j], mlp1_test_score[j]))

        j+=1


```

```python

#I create empty lists where i will save the results of MLP with 1 single layer with different hidden layer sizes
mlps1_train_score=list()
mlps1_test_score=list()

# I initialize the counter variable again
j=0 

# Just printing a title
print(" For the MLP model fit in the SCALED data with a SINGLE layer and different hidden sizes the results are:\n") 

# A loop where the MLP classifier will be fit with the different units for the SCALED data
# I use the units list created before
for i in units: 
        
        mlps1clf=MLPClassifier(hidden_layer_sizes=i, random_state=0).fit(X_train_scaled,y_train)

        # I add to the list each of the scores of the iteration

        mlps1_train_score.append(mlps1clf.score(X_train_scaled,y_train))
        mlps1_train_score.append(mlps1clf.score(X_test_scaled,y_test))
        

        #While looping printing the results 
        print("{}) Hidden size: {}  --> Train Score: {:.3f}   ;   Test Score: {:.3f}".format(j+1,i, mlps1_train_score[j], mlps1_test_score[j]))

        j+=1
```
